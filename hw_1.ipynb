```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("data/weatherHistory.csv")
data.rename({'Formatted Date':'date', 'Temperature (C)':'temp', 'Apparent Temperature (C)':'temp_app',
                    'Precip Type':'weather_type', 'Wind Speed (km/h)':'wind_speed', 'Humidity':'humidity',
                    'Daily Summary':'summary'}, 
                    axis=1, inplace=True)
data.date = pd.to_datetime(data.date, utc=True)
```

1) самый холодный и самый ветренный дни в Сегеде за всю имеющуюся историю

```python
data[data.temp == data.temp.min()]
```

```python
data[data.wind_speed == data.wind_speed.max()]
```

2) сколько раз шел дождь летом 2014 года (все лето)

```python
data[(data['date'] > '2014-06-1') & (data['date'] < '2014-09-01')].sort_values(by="date")\
                                                                  .weather_type\
                                                                  .value_counts()
```

3) Самый ветрянный месяц

```python
months = data[["date", "wind_speed"]].groupby(lambda idx: data.loc[idx].date.month_name())
```

```python
months.aggregate(np.mean).sort_values(by="wind_speed")
```

В среднем, за всю историю самым ветренным является март.

Самый влажный год

```python
years = data[["date", "humidity"]].groupby(lambda idx: data.loc[idx].date.year)
```

```python
years.aggregate(np.mean).sort_values(by="humidity")
```

Самый влажный год: 2005

Неделя с самой переменчивой погодой

```python
weeks = data[["date", "temp", "humidity", "wind_speed"]].groupby(lambda idx: data.loc[idx].date.weekofyear)
```

```python
weeks.aggregate(np.std).sort_values(by=["temp", "humidity", "wind_speed"])
```

Статистически, 6-я неделя года хвастается самой переменчивой погодой
